```python
def histoAU(data: list, bins: int = 30, seed = 1,title: str = 'Semi-Major Axis Dist. for N=1000 & Seed=0'):
    sim = generatettor(simulation=ttor, seed = seed, asteroidnumber=1000)
    auList = []
    for i in range(3,sim.N):
        auList.append(sim.particles[i].a)
    fig, axs = plt.subplots()
    plt.xlabel('Semi-Major Axis (AU)')
    plt.ylabel('Frequency')
    plt.title('Semi-Major Axis Dist. for N=1000 & Seed={}'.format(seed))
    axs.hist(data, bins=bins)
```

```python
for a in range(10):    
    histoAU(auList, seed = a)
```
